```julia
using HTTP, JSON
```

```julia
client_id = "04b07795-8ddb-461a-bbee-02f9e1bf7b46"
scope = "https://graph.microsoft.com/.default%20offline_access"
url_code = "https://login.microsoftonline.com/common/oauth2/v2.0/devicecode"
body_code = "client_id=$client_id&scope=$scope"
r_code = HTTP.post(url_code; body = body_code)
```

```julia
device_code = JSON.parse(String(r_code.body))["device_code"]
url_token = "https://login.microsoftonline.com/common/oauth2/v2.0/token"
body_token = "grant_type=urn:ietf:params:oauth:grant-type:device_code&client_id=$client_id&device_code=$device_code"
open("token_sharepoint", "w") do io
    HTTP.post(url_token; body=body_token, response_stream=io)
end
```

```julia
refresh_token = open(f -> JSON.parse(read(f, String))["refresh_token"], "token_sharepoint", "r")
body_refresh = "client_id=$client_id&grant_type=refresh_token&refresh_token=$refresh_token"
open("token_sharepoint", "w") do io
    HTTP.post(url_token; body=body_refresh, response_stream=io)
end
```

```julia
include("FundsAllocation.jl")
```

```julia
sdf_expenses = FundsAllocation.filter_expenses(FundsAllocation.read_expenses("data.xlsx"))
```

```julia
sdf_funds = FundsAllocation.filter_funds(FundsAllocation.read_funds("data.xlsx"))
```

```julia
expenses_countries = unique(sdf_expenses[!, "CC: Country"])
```

```julia
funds_us = findall(sdf_funds[!, "DONOR CODE"] .== "GUSA02")
```

```julia
sdf_funds[3, "Original Balance"]
```

```julia
for country in expenses_countries
    expenses_country = findall(sdf_expenses[!, "CC: Country"] .== country)
    @constraint(model, c2, sum(x[e, f] * sdf_funds[f, "Original Balance"] * -1 for e in expenses_country, f in funds_us) <= sum(sdf_expenses[expenses_country, "Original Balance"] * 0.66))
end
```
